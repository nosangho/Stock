## 1. 평균 회귀 전략

### 볼린저 밴드
볼린저 밴드는 현재 주가가 상대적으로 높은지 낮은지를 판단할 때 사용하는 보조지표  
볼린저 밴드는 3개 선으로 구성되는데, 중심선인 이동평균선, 상단선과 하단선을 구성하는 표준편차 밴드다.

### 데이터 프레임

```python
df = pdr.get_data_yahoo('AAPL')
df.head()
```

```python
df.describe()
```

### 데이터 가공

```python
price_df = df.loc[:, ['Adj Close']].copy()
price_df.head()
```

```python
# Date가 인덱스로 설정되지 않았을 때 설정하는 방법
price_df.set_index(['Date'], inplace=True)
price_df.head()
```

### 볼린저 밴드 만들기

이동평균선 만들기

```python
price_df['center'] = price_df['Adj Close'].rolling(window = 20).mean()
price_df.iloc[18:25]
```

상한선, 하한선 만들기

```python
price_df['ub'] = price_df['center'] + 2 * price_df['Adj Close'].rolling(window=20).std()
price_df['lb'] = price_df['center'] - 2 * price_df['Adj Close'].rolling(window=20).std()
price_df.iloc[18:25]
```

볼린저 밴드 함수 만들기

```python
n = 20
sigma = 2
def bollinger_band(price_df, n, sigma):
    bb = price_df.copy()
    bb['center'] = price_df['Adj Close'].rolling(window = n).mean()
    bb['ub'] = price_df['center'] + sigma * price_df['Adj Close'].rolling(window=n).std()
    bb['lb'] = price_df['center'] - sigma * price_df['Adj Close'].rolling(window=n).std()
    return bb

bollinger = bollinger_band(price_df, n, sigma)
```

```python
base_date = '2020-05-08'
sample = bollinger.loc[base_date:]
sample.head()
```

거래내역 만들기

```python
book = sample[['Adj Close']].copy()
book['trade'] = ''
book
```

거래내역 함수화

```python
def create_trade_book(sample):
    book = sample[['Adj Close']].copy()
    book['trade'] = ''
    return book
```

### 거래전략

```python
def tradings(sample, book):
    for i in sample.index:
        if sample.loc[i, 'Adj Close'] > sample.loc[i, 'ub']:   # 상단 밴드 이탈 시 동작 안함
            book.loc[i, 'trade'] = ''
        elif sample.loc[i, 'lb'] > sample.loc[i, 'Adj Close']: # 하단 밴드 이탈 시 매수
            if book.shift(1).loc[i, 'trade'] == 'buy':         # 이미 매수 상태라면 매수 유지
                book.loc[i, 'trade'] = 'buy'
            else:
                book.loc[i, 'trade'] = 'buy'                    # 아니면 매수
        elif sample.loc[i, 'ub'] >= sample.loc[i, 'Adj Close'] and sample.loc[i, 'Adj Close'] >= sample.loc[i, 'lb']: #볼린저 밴드 안에 있을 시
            if book.shift(1).loc[i, 'trade'] == 'buy':         # 매수상태면 유지
                book.loc[i, 'trade'] = 'buy'
            else:
                book.loc[i, 'trade'] = ''                       # 동작 안함
    return book
```

```python
book = tradings(sample, book)
book.tail(10)
```

### 전략 수익률

```python
def returns(book):
    # 손익 계산
    rtn = 1.0
    book['return'] = 1
    buy = 0.0
    sell = 0.0
    for i in book.index:
        # long 진입
        if book.loc[i, 'trade'] == 'buy' and book.shift(1).loc[i, 'trade'] == '':
            buy = book.loc[i, 'Adj Close']
            print(f'진입일 : {i}, 진입가격 : {buy}')
        # long 청산
        elif book.loc[i, 'trade'] == '' and book.shift(1).loc[i, 'trade'] == 'buy':
            sell = book.loc[i, 'Adj Close']
            rtn = (sell - buy) / buy + 1 # 손익 계산
            book.loc[i, 'return'] = rtn
            print(f'청산일 : {i}, long 진입가격 : {buy}, long 청산가격 : {sell}, return : {round(rtn, 2)}')
        
        if book.loc[i, 'trade'] == '': # 제로 포지션
            buy = 0.0
            sell = 0.0
            
    acc_rtn = 1.0
    for i in book.index:
        rtn = book.loc[i, 'return']
        acc_rtn = acc_rtn * rtn # 누적수익률 계산
        book.loc[i, 'acc return'] = acc_rtn
    
    print(f'Accunulated return : {round(acc_rtn, 4)}')
    return (round(acc_rtn, 4))
```

## 2. 듀얼 모멘텀 전략

### 듀얼 모멘텀 전략 구현을 위한 절대 모멘텀 전략

- 데이터

```python
import pandas as pd
import numpy as np
import datetime

df = pdr.get_data_yahoo('SPY')
df.head()
```

```python
price_df = df.loc[:, ['Adj Close']].copy()
price_df
```

월말데이터 사용

```python
price_df = price_df.resample(rule='M').last()
price_df['STD_YM'] = price_df.index.map(lambda x : datetime.datetime.strftime(x, '%Y-%m'))
price_df
month_last_df = price_df.copy()
```

```python
# Date가 index가 아닌 경우
price_df.set_index(['Date'], inplace = True) # Date를 index로 만들고 하면 쉬움 (교재에는 엄청 어렵게 해놨음)
price_df = price_df.resample(rule='M').last()
price_df['STD_YM'] = price_df.index.map(lambda x : datetime.datetime.strftime(x, '%Y-%m'))
price_df
```

- 데이터 가공

```python
month_last_df['BF_1M_Adj Close'] = month_last_df.shift(1)['Adj Close']
month_last_df['BF_12M_Adj Close'] = month_last_df.shift(12)['Adj Close']
month_last_df.fillna(0, inplace = True)
month_last_df.head(15)
```

- 포지션 기록

```python
book = price_df.copy()
book['trade'] = ''
book.head()
```

- 거래 실행

```python
# trading 부분
def trading(month_last_df, ticker):
    for x in month_last_df.index:
        signal = ''
        # 절대 모멘텀 계산
        momentum_index = month_last_df.loc[x,'BF_1M_Adj Close'] / month_last_df.loc[x,'BF_12M_Adj Close'] - 1
        flag = True if ((momentum_index > 0.0) and (momentum_index != np.inf) and (momentum_index != -np.nan)) else False and True
        if flag:
            signal = 'buy ' + ticker
        print(f'날짜 : {x}, 모멘텀 인덱스 : {momentum_index}, flag : {flag}, signal : {signal}')
        book.loc[x:, 'trade'] = signal
    return book

book = trading(month_last_df, 'SPY')
```

- 전략 수익률

```python
def returns(book, ticker):
    # 손익 계산
    rtn = 1.0
    book['return'] = 1
    buy = 0.0
    sell = 0.0
    for i in book.index:
        # long 진입
        if book.loc[i, 'trade'] == 'buy ' + ticker and book.shift(1).loc[i, 'trade'] == '':
            buy = book.loc[i, 'Adj Close']
            print(f'진입일 : {i}, 진입가격 : {buy}')
        # 보유 중
        elif book.loc[i, 'trade'] == 'buy ' + ticker and book.shift(1).loc[i, 'trade'] == 'buy ' + ticker:
            current = book.loc[i, 'Adj Close']
            rtn = (current - buy) / buy - 1
            book.loc[i, 'return'] = rtn
        # long 청산
        elif book.loc[i, 'trade'] == '' and book.shift(1).loc[i, 'trade'] == 'buy ' + ticker:
            sell = book.loc[i, 'Adj Close']
            rtn = (sell - buy) / buy + 1 # 손익 계산
            book.loc[i, 'return'] = rtn
            print(f'청산일 : {i}, long 진입가격 : {buy}, long 청산가격 : {sell}, return : {round(rtn, 2)}')
        
        if book.loc[i, 'trade'] == '': # 제로 포지션
            buy = 0.0
            sell = 0.0
            current = 0.0
            
    acc_rtn = 1.0
    for i in book.index:
        if book.loc[i, 'trade'] == '' and book.shift(1).loc[i, 'trade'] == 'buy ' + ticker:
            rtn = book.loc[i, 'return']
            acc_rtn = acc_rtn * rtn # 누적수익률 계산
            book.loc[i, 'acc return'] = acc_rtn

    print(f'Accunulated return : {round(acc_rtn, 4)}')
    return (round(acc_rtn, 4))
```
